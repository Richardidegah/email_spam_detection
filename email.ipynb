```python
import pandas as pd
df = pd.read_csv('/home/user/email/spam.csv', encoding='latin1')
df.head()
```

```python
#group by category and describe
```

```python
df.groupby('v1').describe()
```

```python
#convert v1 columns into numbers (machine laerning cant understand text)
```

```python
#the lamba funtion checks if it's a spam, then return a value of 1, otherwise 0
df['spam']=df['v1'].apply(lambda x :1 if x=='spam' else 0)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.v2, df.spam, test_size=0.25)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
x_train_count = v.fit_transform(x_train.values)
x_train_count.toarray()[:3]
```

```python
#NB x_train_count is the text(emails) converted into a number matrix
#with the code below, the model is trained and ready to make a prediction


from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(x_train_count,y_train)
```

```python
emails = [
    'hello richard, will you be chanced for us to watch football match at the staduim tomorrow?',
    'upto 20% discount just for you on parkaking, exclussive offer. dont miss this regard!'
]
emails_count = v.transform(emails)
model.predict(emails_count)
```

```python
#to measure the accuracy
```

```python
x_test_count = v.transform(x_test)
model.score(x_test_count, y_test)
```

```python
#this email spam detection is better done usinig sklearn pipeline simple API
#NB line 
```

```python
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ("vectorizer", CountVectorizer()),
    ("nb", MultinomialNB())
])
```

```python
clf.fit(x_train, y_train)
```

```python
#accuracy measurement/check the performance of our model

clf.score(x_test, y_test)
```
